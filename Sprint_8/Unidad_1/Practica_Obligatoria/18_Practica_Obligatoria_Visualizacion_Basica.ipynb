### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python

plt.figure(figsize=(16, 12))

# 1. Bar chart of the most populated cities
top_cities = df_cities.nlargest(5, 'population_total')
plt.subplot(2, 2, 1)  # Top-left
colors = ['blue', 'grey', 'salmon', 'olive', 'brown']
bars = plt.bar(top_cities['city'], top_cities['population_total'] / 1000, color=colors, alpha = 0.6)
plt.title('Ciudades ms pobladas de California', fontsize=14)
plt.ylabel('Habitantes (miles)', fontsize=12)
plt.xticks(fontsize=10)

# Adjust the y-axis limit to add more space above the highest bar
max_y = max(top_cities['population_total'] / 1000)  # Get the maximum bar height
plt.ylim(0, max_y + 500)  # Add space by extending the y-axis limit

for bar, pop in zip(bars, top_cities['population_total']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 60, f'{pop / 1000:.3f}', 
             ha='center', color='black', fontsize=10, fontname='Arial')



# 2. Histogram of city areas
plt.subplot(2, 2, 2)  # Top-right
plt.hist(df_cities['area_total_km2'].dropna(), bins=30, alpha=0.7, color='skyblue', label='Hist Area')

plt.xlabel(r'$\mathit{Superficie}$ (km)', fontsize=12)
plt.ylabel('Nmero de ciudades', fontsize=12, fontstyle='italic')
plt.legend()

# 3. Scatter plot of population density vs water area
plt.subplot(2, 2, 3)  # Bottom-left
population_density = df_cities['population_total'] / (df_cities['area_water_km2'] + 1e-9)  # Avoid division by zero
plt.scatter(df_cities['area_water_km2'], population_density, alpha=0.7, color='green')
plt.title('Poblacin por kilmetro cuadrado de agua', fontsize=14)
plt.xlabel('rea de agua en la ciudad (km)', fontsize=12)
plt.ylabel('Poblacin (miles / km)', fontsize=12)
# Highlight outliers
outliers = population_density > 1000
for i in np.where(outliers)[0]:
    plt.annotate('Outliers', (df_cities['area_water_km2'].iloc[i], population_density.iloc[i]),
                 textcoords="offset points", xytext=(5, 5), ha='center')

# 4. Box plot of elevations
plt.subplot(2, 2, 4)  # Bottom-right
plt.boxplot(df_cities['elevation_m'].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='red'))
plt.title('Diagrama de caja de la Altitud de las ciudades de California', fontsize=14)
plt.xlabel('Altitud (m)', fontsize=12)

# Adjust layout
plt.tight_layout()
plt.show()
```
