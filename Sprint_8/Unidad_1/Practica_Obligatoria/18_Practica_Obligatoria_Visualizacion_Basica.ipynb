### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python

plt.figure(figsize=(16, 12))

# 1. Bar chart of the most populated cities
top_cities = df_cities.nlargest(5, 'population_total')
plt.subplot(2, 2, 1)  # Top-left
colors = ['blue', 'grey', 'salmon', 'olive', 'brown']
bars = plt.bar(top_cities['city'], top_cities['population_total'] / 1000, color=colors, alpha = 0.6)
plt.title('Ciudades ms pobladas de California', fontsize=14)
plt.ylabel('Habitantes (miles)', fontsize=12)
plt.xticks(fontsize=10)

# Adjust the y-axis limit to add more space above the highest bar
max_y = max(top_cities['population_total'] / 1000)  # Get the maximum bar height
plt.ylim(0, max_y + 500)  # Add space by extending the y-axis limit

for bar, pop in zip(bars, top_cities['population_total']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 60, f'{pop / 1000:.3f}', 
             ha='center', color='black', fontsize=10, fontname='Arial')



# 2. Histogram of city areas
plt.subplot(2, 2, 2)  # Top-right
plt.hist(df_cities['area_total_km2'].dropna(), bins=30, alpha=0.7, color='skyblue', label='Hist Area')

plt.xlabel(r'$\mathit{Superficie}$ (km)', fontsize=12)
plt.ylabel('Nmero de ciudades', fontsize=12, fontstyle='italic')
plt.legend()

# Define font properties for customization
font_main = font_manager.FontProperties(family='serif', style='italic', size=12)
font_annotation = font_manager.FontProperties(family='Arial', style='italic', weight='bold', size=10)

# 3. Scatter plot of water area vs population

plt.subplot(2, 2, 3)  # Bottom-left position in the grid

# Scale the population to thousands (miles habitantes)
population_in_thousands = df_cities['population_total'] / 1000

# Scatter plot
plt.scatter(df_cities['area_water_km2'], population_in_thousands, alpha=0.7, color='green')
plt.title('Poblacin por kilmetro cuadrado de agua', fontsize=14, fontproperties=font_main)
plt.xlabel('rea de agua en la ciudad (km)', fontsize=12, fontproperties=font_main)
plt.ylabel('Poblacin (miles habitantes)', fontsize=12, fontproperties=font_main)

# Highlight the outliers
outliers = df_cities[df_cities['population_total'] > 2000000]  # Example threshold for large populations

# Annotate the outliers dynamically
for index, row in outliers.iterrows():
    plt.annotate('Outliers',
                 (row['area_water_km2'], row['population_total'] / 1000),  # Point to annotate
                 textcoords="offset points",
                 xytext=(-50, -40),  # Adjusted position for text
                 ha='center',
                 fontproperties=font_annotation,
                 arrowprops=dict(facecolor='black', arrowstyle='->'))

# Adjust axis ranges for better clarity with padding
x_padding = df_cities['area_water_km2'].max() * 0.04 
y_padding = population_in_thousands.max() * 0.04  # 
plt.xlim(-x_padding, df_cities['area_water_km2'].max() + x_padding)  # Add padding to x-axis
plt.ylim(-y_padding, population_in_thousands.max() + y_padding)  # Add padding to y-axis

# Offset the axes spines slightly
ax = plt.gca()
ax.spines['left'].set_position(('outward', 10))  # Move y-axis spine outward
ax.spines['bottom'].set_position(('outward', 10))  # Move x-axis spine outward

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.6)

# Format the y-axis ticks to show integers (e.g., 0, 500, 1000)
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x)}'))

plt.tight_layout()

# 4. Box plot of elevations
plt.subplot(2, 2, 4)  # Bottom-right
plt.boxplot(df_cities['elevation_m'].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='red'))
plt.title('Diagrama de caja de la Altitud de las ciudades de California', fontsize=14)
plt.xlabel('Altitud (m)', fontsize=12)

# Adjust layout
plt.tight_layout()
plt.show()
```
